```python
import os 
import numpy as np 
```

```python
def slash(n=30):
    print("-" * n)

def check_setting(dataset_name):
    print(f'dataset: {dataset_name}')
    train_path = os.path.join(dataset_name, "train.npy")
    test_path = os.path.join(dataset_name, "test.npy")
    test_label_path = os.path.join(dataset_name, "test_label.npy")

    train = np.load(train_path)
    test = np.load(test_path)
    test_label = np.load(test_label_path)
    # print(train)
    n_features = train.shape[1]
    print(f'n features: {n_features}')
    print(train.shape, test.shape, test_label.shape)

    print(f'n sample train+test: {train.shape[0] + test.shape[0]}')
    print(sum(test_label))

    slash()

dataset_names = os.listdir()
# print("dataset_names: ", dataset_names)

# only keep folder
dataset_names = [name for name in dataset_names if os.path.isdir(name)]
print("dataset_names: ", dataset_names)

for dataset_name in dataset_names:
    check_setting(dataset_name)
    # break 
```
